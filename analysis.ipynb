```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv('bio_2.csv',encoding='latin1')
```

```python
df
```

```python
df.rename(columns={'WaterÊ': 'Water',
                   'PondWaterÊ': 'PondWater',
                  'TeaÊ':'Tea'},
          inplace=True, errors='raise')
```

```python
df
```

```python
df=df.drop(columns=['Unnamed: 0'])
```

```python
df
```

```python
sns.set_style("darkgrid");
plt.plot(df);
plt.title("growth of my plants")
plt.xticks(np.arange(0,12));
plt.legend( ('Water', 'SaltWater', 'PondWater','Tea'))
plt.xlabel('day')
plt.ylabel('size in cm')
```
